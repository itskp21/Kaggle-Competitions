```python
train = pd.read_csv("/kaggle/input/spaceship-titanic/train.csv")
test = pd.read_csv("/kaggle/input/spaceship-titanic/test.csv")
test_ids = test["PassengerId"]
train
```

```python
train = train.drop(['PassengerId','Name'],axis = 1)
```

```python
train.dtypes
```

```python
train.isna().sum()
```

```python
train.select_dtypes(include="float")
```

```python
train[['VIP', 'CryoSleep', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']] = train[['VIP', 'CryoSleep', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']].fillna(value=0)
train.isnull().sum().sort_values(ascending=False)
```

```python
label = "Transported"
train[label] = train[label].astype(float)
```

```python
train[["Deck", "Cabin_num", "Side"]] = train["Cabin"].str.split("/", expand=True)
```

```python
train = train.drop('Cabin', axis=1)
```

```python
train
```

```python
train.dtypes
```

```python
train.isna().sum()
```

```python
train['HomePlanet'].fillna('missing',inplace = True)
train['Destination'].fillna('missing',inplace = True)
train['Age'].fillna(train['Age'].median(),inplace = True)
train['RoomService'].fillna(train['RoomService'].median(),inplace = True)
train['Deck'].fillna('missing',inplace = True)
train['Cabin_num'].fillna('missing',inplace = True)
train['Side'].fillna('missing',inplace = True)
```

```python
train.isna().sum()
```

```python
train
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

cols = ['HomePlanet','CryoSleep','Destination','VIP','Deck','Cabin_num','Side']

for col in cols:
    train[col]= le.fit_transform(train[col])
```

```python
train
```

```python
train.dtypes
```

```python
x = train.drop('Transported',axis =1)
y = train['Transported']
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,
                                                y,
                                                test_size = 0.2,
                                                random_state = 42)
```

```python
len(x_train),len(y_train)
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
clf.fit(x_train,y_train)
clf.score(x_test,y_test)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
clf = GradientBoostingClassifier()
clf.fit(x_train,y_train)
clf.score(x_test,y_test)
```

```python
import lightgbm
from lightgbm import LGBMClassifier
clf = LGBMClassifier()
clf.fit(x_train,y_train)
clf.score(x_test,y_test)
```

```python
predictions = clf.predict(x_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test,predictions)
```
